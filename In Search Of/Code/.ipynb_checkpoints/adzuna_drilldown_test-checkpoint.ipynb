# Adzuna Drill-Down Test — watch the code work

This notebook loads the **real functions** from `In_Search_Of.py` (it reads the source file, so you're watching the exact pushed code — not a copy) and runs them step-by-step against the Emory "Environmental Services" posting.

It answers the two open issues directly:
1. **Apply link still points at adzuna.com** — Steps 1 & 3 show whether the drill-down resolves the *true* employer URL (emory.jobs).
2. **The 222-day-old posting survives** — Steps 1, 2 & 4 show whether the age gate fetches the real `datePosted` and drops it.

**Key insight to watch for:** locally (your residential IP) the fetches succeed, so you'll see it work. In GitHub Actions (datacenter IP) Adzuna blocks ~half the fetches — that's the production failure. **Step 5** tests whether the Adzuna *API* `created` date can fix this with no website fetching at all.

## Setup — load the real helpers from In_Search_Of.py

We slice out the fetch/age/resolve/enrich region of the source and `exec` it into a namespace, stubbing only `update_page` (so we can watch the drop decision without writing to Notion).

```python
import re, json, time, datetime, requests, os
from urllib.parse import urlparse
from pathlib import Path

# Edit this if your checkout is elsewhere.
ISO_PATH = Path("/Users/couch2coders/Documents/GitHub/newsletters/In Search Of/Code/In_Search_Of.py")
src = ISO_PATH.read_text()

# Grab the real helper + enrich region (everything from the Adzuna regexes
# down to, but not including, main()).
start = src.index("_ADZUNA_DETAILS_RE = re.compile")
end   = src.index("def main(")
slice_src = src[start:end]

# Stub Notion writes so we can watch the logic without touching the DB.
def update_page(page_id, properties=None):
    print(f"      [stub] update_page({page_id}) -> {list((properties or {}).keys())}")

ns = {"re": re, "json": json, "time": time, "datetime": datetime,
      "requests": requests, "urlparse": urlparse, "print": print,
      "update_page": update_page}
exec(compile(slice_src, str(ISO_PATH), "exec"), ns)

print("loaded functions:", sorted(k for k in ns if callable(ns[k]) and not k.startswith("__")))
print("MAX_LISTING_AGE_DAYS =", ns["MAX_LISTING_AGE_DAYS"])
try:
    import curl_cffi; print("curl_cffi:", curl_cffi.__version__, "(browser impersonation available)")
except ImportError:
    print("curl_cffi NOT installed — falls back to plain requests (likely blocked in CI)")
```

## Step 1 — fetch the Adzuna details page

`_fetch` (persistent curl_cffi Session) and `_fetch_details` (single fetch → datePosted + employer + the `/land/ad` apply-redirect link).

```python
EMORY_URL = ("https://www.adzuna.com/details/5763134132?se=Gqmtky1s8RGrKszqFbIEfQ"
             "&utm_medium=api&utm_source=893fec3f&v=1EC413E219C5C6F1EE241F3B1AB3C84C8611E059")

final_url, html = ns["_fetch"](EMORY_URL)
print("fetch final URL :", final_url[:80])
print("HTML length     :", len(html), "(0 = blocked/non-200)")

details = ns["_fetch_details"](EMORY_URL)
print("date_posted     :", details.get("date_posted"))
print("employer        :", details.get("employer"))
print("land (redirect) :", (details.get("land") or "")[:80])
```

## Step 2 — age check (should DROP — ~222 days old)

This is the exact comparison the age gate makes.

```python
today  = datetime.date.today()
posted = ns["_parse_iso_date"](details.get("date_posted", ""))
age    = (today - posted).days if posted else None
print("posted date :", posted)
print("today       :", today)
print("age (days)  :", age)
print("max allowed :", ns["MAX_LISTING_AGE_DAYS"])
print("DECISION    :", "DROP (stale)" if (age is not None and age > ns["MAX_LISTING_AGE_DAYS"])
      else ("keep" if age is not None else "keep (no date — fetch blocked?)"))
```

## Step 3 — resolve the TRUE apply URL

Chain: `adzuna /land/ad` → (meta-refresh / JS) aggregator `de.jobsyn.org` → employer site `emory.jobs`. The land page does **not** HTTP-redirect, so we read the client-side redirect target and follow it.

```python
land = details.get("land")
print("land (adzuna)      :", (land or "")[:80], "\n")

_, land_html = ns["_fetch"](land)
dest = ns["_meta_redirect"](land_html)
print("meta/JS redirect   :", (dest or "(none found)")[:90])

final_apply, _ = ns["_resolve_apply_url"](land)
print("FINAL apply URL    :", final_apply[:100])
print("left adzuna.com?   :", not ns["_is_adzuna_host"](final_apply), "  <-- True means it's the real employer URL")
```

## Step 4 — full `enrich_adzuna_rows` on a fake pool

Runs the *actual* two-phase enrichment (Notion writes stubbed). Watch: the age-gate log lines, Emory getting marked rejected + removed from the pool, and the non-adzuna row passing through untouched.

```python
pool = [
    {"notion_page_id": "EMORY", "employer": "Emory Healthcare/Emory University",
     "scraped_snippet": "existing snippet", "job_listings_url": EMORY_URL},
    {"notion_page_id": "OTHER", "employer": "Some Non-Adzuna Co",
     "scraped_snippet": "existing", "job_listings_url": "https://example.com/careers/job-1"},
]
print("BEFORE:")
for r in pool: print("  ", r["notion_page_id"], "|", r["job_listings_url"][:45])
print("\n--- enrich_adzuna_rows running ---")
ns["enrich_adzuna_rows"](pool)
print("\nAFTER:")
for r in pool: print("  ", r["notion_page_id"], "|", r["job_listings_url"][:55])
print("\nEmory still in pool?:", any(r["notion_page_id"] == "EMORY" for r in pool),
      "  <-- False means the stale posting was dropped")
```

## Step 5 (optional) — is the Adzuna API `created` date honest?

**This is the decisive test for the CI failure.** If Adzuna's API reports the *true* post date in `created`, we can drop stale postings at scrape time with **zero website fetching** (immune to the IP block). If `created` is a recent re-index date, scrape-time filtering can't catch them and we need another path.

Needs `ADZUNA_APP_ID` / `ADZUNA_APP_KEY` in the environment. Set them, then run this cell.

```python
APP_ID  = os.environ.get("ADZUNA_APP_ID", "")
APP_KEY = os.environ.get("ADZUNA_APP_KEY", "")
if not (APP_ID and APP_KEY):
    print("Set ADZUNA_APP_ID and ADZUNA_APP_KEY in the environment, then re-run this cell.")
else:
    params = {"app_id": APP_ID, "app_key": APP_KEY, "where": "Marietta, GA",
              "distance": 24, "results_per_page": 50, "max_days_old": 21,
              "sort_by": "date", "category": "healthcare-nursing-jobs"}
    r = requests.get("https://api.adzuna.com/v1/api/jobs/us/search/1", params=params, timeout=25)
    data = r.json()
    print(f"API returned {len(data.get('results', []))} results (max_days_old=21)\n")
    print(f"{'created':12} {'age':>5}  employer / title")
    print("-" * 70)
    today = datetime.date.today()
    for j in data.get("results", []):
        c = (j.get("created") or "")[:10]
        try:
            age = (today - datetime.date.fromisoformat(c)).days
        except Exception:
            age = "?"
        comp  = ((j.get("company") or {}).get("display_name") or "")[:28]
        title = (j.get("title") or "")[:34]
        flag  = "  <-- EMORY/ENV" if ("emory" in comp.lower() or "environmental" in title.lower()) else ""
        print(f"{c:12} {age!s:>5}  {comp:28} {title}{flag}")
    print("\nInterpretation:")
    print(" - If the Emory/Environmental row shows created ~2025-11-11 (~222d): the API created")
    print("   IS the true date -> scrape-time filtering catches it (and max_days_old is buggy).")
    print(" - If it shows a recent created (<21d): the API re-indexed it -> scrape-time filtering")
    print("   can't catch it; only the website datePosted can, so the drill-down must succeed.")
```

### How to read the results

- **Steps 1–4 all work (Emory dropped, apply URL = emory.jobs):** the code is correct; the only reason it fails in production is Adzuna blocking GitHub's IP. → fix = make the fetch reliable in CI **or** use the API path (Step 5).
- **Step 5 shows Emory's `created` ≈ 222 days:** best fix — filter at scrape time on `created`, no website fetch, immune to the block.
- **Step 5 shows Emory's `created` recent:** the API hides the real age → we must get the website `datePosted` reliably (residential proxy, or retry/session tuning), since nothing else has the true date.

Swap `EMORY_URL` for any other `adzuna.com/details/...` URL to test a different posting.
